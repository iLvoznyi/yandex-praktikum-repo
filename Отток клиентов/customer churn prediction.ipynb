## Исследование задачи

### Выделение целевого признака и разбиение данных на выборки

Проблему дисбаланса классов выявили в первом шаге - целевой признак несбалансирован соотношение примерно 80 на 20, перейдём к обучению модели без учёта дисбаланса

Выделим целевой признак и разобьём исходные данные на обучающую выборку и временную выборку в соотношении 60:40. Затем мы разделяем временную выборку на валидационную и тестовую выборки в соотношении 50:50.

```python
features = data.drop(columns=['Exited'])
target = data['Exited']

train_data, temp_data, train_target, temp_target = train_test_split(features, target, test_size=0.4, random_state=12345)
val_data, test_data, val_target, test_target = train_test_split(temp_data, temp_target, test_size=0.5, random_state=12345)
```

Кодирирование столбцов Geography Gender с помошью OneHotEncoder

```python
encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
train_data_encoded = encoder.fit_transform(train_data[['Geography', 'Gender']]).toarray()
encoded_columns = encoder.get_feature_names_out(['Geography', 'Gender'])
train_data_encoded_df = pd.DataFrame(train_data_encoded, columns=encoded_columns, index=train_data.index)
train_data.drop(columns=['Geography', 'Gender'], inplace=True)
train_data = pd.concat([train_data, train_data_encoded_df], axis=1)

val_data_encoded = encoder.transform(val_data[['Geography', 'Gender']]).toarray()
val_data_encoded_df = pd.DataFrame(val_data_encoded, columns=encoded_columns, index=val_data.index)
val_data.drop(columns=['Geography', 'Gender'], inplace=True)
val_data = pd.concat([val_data, val_data_encoded_df], axis=1)

test_data_encoded = encoder.transform(test_data[['Geography', 'Gender']]).toarray()
test_data_encoded_df = pd.DataFrame(test_data_encoded, columns=encoded_columns, index=test_data.index)
test_data.drop(columns=['Geography', 'Gender'], inplace=True)
test_data = pd.concat([test_data, test_data_encoded_df], axis=1)
```

Масштабирование с помощью StandardScaler

```python
scaler = StandardScaler()
columns_to_scale = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']
train_data[columns_to_scale] = scaler.fit_transform(train_data[columns_to_scale])
val_data[columns_to_scale] = scaler.transform(val_data[columns_to_scale])
test_data[columns_to_scale] = scaler.transform(test_data[columns_to_scale])
```

```python
display("Размер обучающей выборки:", train_data.shape)
display("Размер валидационной выборки:", val_data.shape)
display("Размер тестовой выборки:", test_data.shape)
```

### Обучение моделей

#### Случайный лес

```python
best_f1_rf = 0
best_auc_rf = 0
best_params_rf = {}
for n_estimators in range(1, 40):
    for max_depth in range(5, 20):
        model_rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, random_state=12345)
        model_rf.fit(train_data, train_target)
        predictions_rf = model_rf.predict(val_data)
        accuracy_rf = accuracy_score(val_target, predictions_rf)
        f1_rf = f1_score(val_target, predictions_rf)
        probabilities_rf = model_rf.predict_proba(val_data)
        probabilities_one_rf = probabilities_rf[:, 1]
        roc_auc_rf = roc_auc_score(val_target, probabilities_one_rf)
        print('Количество деревьев:', n_estimators, 'Глубина дерева:', max_depth)
        print('F1-мера:', f1_rf)
        print()
        if f1_rf > best_f1_rf:
            best_f1_rf = f1_rf
            best_auc_rf = roc_auc_rf
            best_params_rf = {
                'n_estimators': n_estimators,
                'max_depth': max_depth
            }

print('Лучшие гиперпараметры для Случайного леса:', best_params_rf)
print('Лучшая F1-мера:', best_f1_rf)
print('Лучшая AUC-ROC:', best_auc_rf)
print()
```

#### Логическая регрессия

```python
# Логистическая регрессия
best_f1_lr = 0
best_auc_lr = 0
best_params_lr = {}

for C in np.arange(0.01, 1.0, 0.01):
    model_lr = LogisticRegression(C=C, random_state=12345)
    model_lr.fit(train_data, train_target)
    predictions_lr = model_lr.predict(val_data)
    accuracy_lr = accuracy_score(val_target, predictions_lr)
    f1_lr = f1_score(val_target, predictions_lr)
    probabilities_lr = model_lr.predict_proba(val_data)
    probabilities_one_lr = probabilities_lr[:, 1]
    roc_auc_lr = roc_auc_score(val_target, probabilities_one_lr)
    if f1_lr > best_f1_lr:
        best_f1_lr = f1_lr
        best_auc_lr = roc_auc_lr
        best_params_lr = {'C': C}

    print('Гиперпараметры:', {'C': C})
    print('F1-мера:', f1_lr)
    print()

print('Лучшие гиперпараметры для Логистической регрессии:', best_params_lr)
print('Лучшая F1-мера:', best_f1_lr)
print('Лучшая AUC-ROC:', best_auc_lr)
print()
```

#### Дерево решений

```python
# Дерево решений
best_f1_dt = 0
best_auc_dt = 0
best_params_dt = {}

for max_depth in range(5, 20):
    for min_samples_split in range(2, 10):
        for min_samples_leaf in range(1, 10):
            for criterion in ['gini', 'entropy']:
                model_dt = DecisionTreeClassifier(max_depth=max_depth, min_samples_split=min_samples_split,
                                                  min_samples_leaf=min_samples_leaf, criterion=criterion,
                                                  random_state=12345)
                model_dt.fit(train_data, train_target)
                predictions_dt = model_dt.predict(val_data)
                accuracy_dt = accuracy_score(val_target, predictions_dt)
                f1_dt = f1_score(val_target, predictions_dt)
                probabilities_dt = model_dt.predict_proba(val_data)
                probabilities_one_dt = probabilities_dt[:, 1]
                roc_auc_dt = roc_auc_score(val_target, probabilities_one_dt)
                print('Гиперпараметры:', {
                    'max_depth': max_depth,
                    'min_samples_split': min_samples_split,
                    'min_samples_leaf': min_samples_leaf,
                    'criterion': criterion
                })
                print('F1-мера:', f1_dt)
                print()
                if f1_dt > best_f1_dt:
                    best_f1_dt = f1_dt
                    best_auc_dt = roc_auc_dt
                    best_params_dt = {
                        'max_depth': max_depth,
                        'min_samples_split': min_samples_split,
                        'min_samples_leaf': min_samples_leaf,
                        'criterion': criterion
                    }

print('Лучшие гиперпараметры для Дерева решений:', best_params_dt)
print('Лучшая F1-мера:', best_f1_dt)
print('Лучшая AUC-ROC:', best_auc_dt)
```

## Борьба с дисбалансом

Для решения дисбаланса воспльзуемся двумя методами: увеличением выборки и class weighting. А также посмотрим, что будет при применении двух методов одновременно

### Метод upsampling

Увеличим количество значений 1 для целевого признка равным количеству значений 0, с помощью собственной функции upsample и передадим ей значение 4, тк примерная разница признкаво 4 раза.

```python
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    return features_upsampled, target_upsampled

upsampled_data, upsampled_target = upsample(train_data, train_target, 4)
```
