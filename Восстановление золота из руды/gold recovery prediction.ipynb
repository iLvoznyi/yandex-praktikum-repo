# Описание данных
**Технологический процесс** 
 * Rougher feed — исходное сырье
 * Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
    - Xanthate — ксантогенат (промотер, или активатор флотации);
    - Sulphate — сульфат (на данном производстве сульфид натрия);
    - Depressant — депрессант (силикат натрия).
 * Rougher process (англ. «грубый процесс») — флотация
 * Rougher tails — отвальные хвосты
 * Float banks — флотационная установка
 * Cleaner process — очистка
 * Rougher Au — черновой концентрат золота
 * Final Au — финальный концентрат золота

**Параметры этапов**
 * air amount — объём воздуха
 * fluid levels — уровень жидкости
 * feed size — размер гранул сырья
 * feed rate — скорость подачи
 
# Наименование признаков

Наименование признаков должно быть такое:

[этап].[тип_параметра].[название_параметра]
Пример: rougher.input.feed_ag

Возможные значения для блока [этап]:

 * rougher — флотация
 * primary_cleaner — первичная очистка
 * secondary_cleaner — вторичная очистка
 * final — финальные характеристики

Возможные значения для блока [тип_параметра]:

 * input — параметры сырья
 * output — параметры продукта
state — параметры, характеризующие текущее состояние этапа
calculation — расчётные характеристики

### Проверка верности вычисления эффективности обогащения

```python
F = data_train['rougher.input.feed_au']
C = data_train['rougher.output.concentrate_au']
T = data_train['rougher.output.tail_au']
recovery = C*(F-T)/(F*(C-T))*100
print("Значение MAE:", mean_absolute_error(data_train['rougher.output.recovery'], recovery))
```

Значение MAE близко к нулю, что говорит верности вычисления, а также о достоверность и точности данных

## Модель

Функции для рассчёта sMAPE

```python
def sMAPE(y, y_pred):
    numerator = np.abs(y - y_pred)
    denominator = (np.abs(y) + np.abs(y_pred)) / 2
    smape = np.mean(numerator / denominator) * 100
    return smape
smape = make_scorer(sMAPE, greater_is_better=False)
```

Разделение выборки на тестовую и тренировочную, а также разделение на обучающие признаки и целевые

```python
features_columns = list(data_test.columns)[:-2]
target_columns = list(reversed(list(data_test.columns)[-2:]))
features_train = data_train[features_columns]
target_train = data_train[target_columns]
features_test = data_test[features_columns]
target_test = data_test[target_columns]
```

### Линейная регресия

```python
state = np.random.RandomState(12345)
print("Линейная регресия")
model = LinearRegression()
scores_r = cross_val_score(model, features_train, target_train.iloc[:, 0], scoring=smape, cv=5) 
scores_f = cross_val_score(model, features_train, target_train.iloc[:, 1], scoring=smape, cv=5) 
final_score = -1*pd.Series(0.25*scores_r+0.75*scores_f).mean()
print('Средняя оценка качества модели:', final_score)
```

### Случайный лес

```python
# Случайный лес
print("3. Случайный лес")
forest_model = RandomForestRegressor(random_state=12345)

parameters = {
    'max_depth': [x for x in [2,5]],
    'n_estimators': [x for x in [100,150]]
}

best_forest_final_score = np.inf
best_forest_params = {}

for max_depth in parameters['max_depth']:
    for n_estimators in parameters['n_estimators']:
        forest_model.set_params(max_depth=max_depth, n_estimators=n_estimators)
        forest_model.fit(features_train, target_train)
        print(1)
        forest_scores_r = cross_val_score(forest_model, features_train, target_train.iloc[:, 0], scoring=smape, cv=5)
        print(1)
        forest_scores_f = cross_val_score(forest_model, features_train, target_train.iloc[:, 1], scoring=smape, cv=5)
        forest_final_score = -1 * pd.Series(0.25 * forest_scores_r + 0.75 * forest_scores_f).mean()
        
        print("Параметры модели: max_depth =", max_depth, "n_estimators =", n_estimators)
        print("Значение itog_sMAPE:", forest_final_score)

        if forest_final_score < best_forest_final_score:
            best_forest_final_score = forest_final_score
            best_forest_params = {'max_depth': max_depth, 'n_estimators': n_estimators}

print("Лучшие параметры:", best_forest_params)
print("Лучшее значение itog_sMAPE:", best_forest_final_score)
```

### Дерево решений

```python
print("1. Дерево решений для регрессии")
tree_model = DecisionTreeRegressor(random_state=12345)

tree_params = {'max_depth': range(1,3,1)}

best_tree_final_score = np.inf
best_tree_params = {}

for max_depth in tree_params['max_depth']:
    tree_model.set_params(max_depth=max_depth)
    tree_model.fit(features_train, target_train)

    tree_scores_r = cross_val_score(tree_model, features_train, target_train.iloc[:, 0], scoring=smape, cv=5)
    tree_scores_f = cross_val_score(tree_model, features_train, target_train.iloc[:, 1], scoring=smape, cv=5)
    tree_final_score = -1 * pd.Series(0.25 * tree_scores_r + 0.75 * tree_scores_f).mean()

    print("Параметры модели: max_depth =", max_depth)
    print("Значение itog_sMAPE:", tree_final_score)

    if tree_final_score < best_tree_final_score:
        best_tree_final_score = tree_final_score
        best_tree_params = {'max_depth': max_depth}

print("Лучшие параметры:", best_tree_params)
print("Лучшее значение itog_sMAPE:", best_tree_final_score)
```
